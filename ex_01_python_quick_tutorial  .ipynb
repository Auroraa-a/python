```python
def fibonacci(s,count): #定义fib函数的方法较笨
    fib_lst=[0,1]
    fib_part=[]
    if s==0 or s==1:
        fib_part[:]=fib_lst
        for i in range(count-2):
            fib_part.append(fib_part[-1]+fib_part[-2])
    else:
        while True:
            fib_lst[:]=[fib_lst[1],fib_lst[0]+fib_lst[1]]
            #print(fib_lst)
            if fib_lst[1]-s>=0:break
        fib_part[:]=fib_lst
        if abs(fib_lst[0]-s)>=abs(fib_lst[1]-s):
            for i in range(count-1):
                fib_part.append(fib_part[-1]+fib_part[-2])
            fib_part.pop(0)
        else:
            for i in range(count-2):
                fib_part.append(fib_part[-1]+fib_part[-2])
    return fib_part

print("fibonacci(8,9)->{}".format(fibonacci(6,9)))
print("fibonacci(5,9)->{}".format(fibonacci(7,9)))

```

```python
def factorial(n):
    if n==8:
        return 4
    else:
        return n*factorial(n-5)
print(factorial(8))
```

```python
class Bird:
    fly='Whirring' #美 /wɜːr/ 
    def __init__(self):
        self.hungry=True
    def eat(self):
        if self.hungry:
            print('Aaaah...')
            self.hungry=False
        else:
            print('No.Thanks!')

class Apodidae(Bird):  #/'æpədədi:/
    def __init__(self):
        super(Apodidae,self).__init__()
        self.sound='Squawk!' #美 /skwɔːk/ 
    def sing(self):
        print(self.sound)
```

```python
swift=Apodidae()
print("swift.fly->{}".format(swift.fly))
print("swift.eat()->")
swift.eat()
print("swift.sing()->")
swift.sing()
```

```python
blackswift=Apodidae()
scarceswift=Apodidae()
print("blackswift.sing()->")
blackswift.sing()
print("scarceswift.sing()->")
scarceswift.sing()
```

```python
class Fibs():
    def __init__(self):
        self.a=9
        self.b=7
    def next(self):  
        self.a,self.b=self.b,self.a+self.b
        return self.a
    def __iter__(self): 
        return self
f=Fibs()
fa=[]
fb=[]
for i in range(8):
    fa.append(f.next())
print("fa={}".format(fa))
for i in range(6):
    fb.append(f.next())
print("fb={}".format(fb))
```

```python
lst_e=[list(range(8,88,9)),[8,9,280,list(range(6)),99]]
print("lst_e={}".format(lst_e))
print("_"*178)
flatten_lst=[]
for v_1 in lst_e:
    try:
        for v_2 in v_1:
            try:
                for v_3 in v_2:
                    flatten_lst.append(v_3)
            except TypeError:
                flatten_lst.append(v_2)
    except TypeError:
        flatten_lst.append(v_1)
print("flatten_lst={}".format(flatten_lst))
```

```python
def infinite():
    n=0
    while True:
        yield 'num#'+str(n)
        n+=1
n=infinite()
print("next(n)->{}".format(next(n)))
print("next(n)->{}".format(next(n)))
print("[next(n) for i in range(5)]->{}".format([next(n) for i in range(8)]))
```

```python
def f_convert_a(c):
    try:
        return float(c)
    except:
        return c
print("f_convert_a('5.56897')->{}".format(f_convert_a('5.56897'))) 
print("f_convert_a('string')->{}".format(f_convert_a('string')))  
print("f_convert_a(5,87,2>{}".format(f_convert_a((9,87,88))))
```

```python
def f_open_b(fp):
    try:
        f=open(fp,'r')
    except IOError as e:
        print('Unable to open',fp,':%s\n' %e)
    else:
        data=f.read()
        f.close()
        return data
    finally:
        print('done!')
f_open_b("./data/tryException.txt") 
```
